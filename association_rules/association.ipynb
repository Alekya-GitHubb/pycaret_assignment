```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/groceries.csv"

# Read as raw text, not CSV
response = requests.get(url)
data = response.text.splitlines()

# Convert the list of transactions into a pandas DataFrame
data = pd.DataFrame({'items': data})

display(data.head())
```

```python
transactions = data["items"].apply(lambda x: x.split(","))
transactions.head()
```

```python
!pip install mlxtend efficient-apriori --quiet
```

```python
from efficient_apriori import apriori

itemsets, rules = apriori(transactions, min_support=0.01, min_confidence=0.3)

print("✅ Rules Generated:")
for r in rules[:10]:   # show first 10
    print(r)
```

```python
data.to_csv("groceries_transactions_raw.csv", index=False)
print("✅ groceries_transactions_raw.csv saved")
```

```python
df_expanded = data["items"].str.split(",", expand=True)
df_expanded.to_csv("groceries_transactions_expanded.csv", index=False)
print("✅ groceries_transactions_expanded.csv saved")
```

```python
import pandas as pd

rules_df = []

for r in rules:
    rules_df.append({
        "antecedent": ','.join(r.lhs),
        "consequent": ','.join(r.rhs),
        "support": r.support,
        "confidence": r.confidence,
        "lift": r.lift
    })

rules_df = pd.DataFrame(rules_df)
rules_df.to_csv("association_rules.csv", index=False)

print("✅ association_rules.csv saved")
```
