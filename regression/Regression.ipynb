```python
%%bash
micromamba run -n pycaret38 python - << 'EOF'
from pycaret.regression import *
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = load_diabetes(as_frame=True).frame

s = setup(data, target="target", session_id=42, normalize=True, silent=True, html=False)
best = compare_models()
model = create_model(best)
tuned = tune_model(model)

# ---- FIX: Handle Linear models vs Tree models ----
try:
    importances = tuned.feature_importances_
    cols = tuned.feature_names_in_
except:
    # Use coefficients instead (linear models)
    try:
        importances = np.abs(tuned.coef_)
        cols = tuned.feature_names_in_
    except:
        print("⚠️ This model does not support feature importance or coefficients.")
        importances = None

# Plot if available
if importances is not None:
    plt.figure(figsize=(6,4))
    plt.barh(cols, importances)
    plt.title("Feature Importance")
    plt.tight_layout()
    plt.savefig("feature_importance_custom.png")
    plt.close()
    print("✅ Saved: feature_importance_custom.png")
else:
    print("⚠️ No feature importance plot generated")
EOF

```

```python
%%bash
micromamba run -n pycaret38 python - << 'EOF'
from pycaret.regression import *
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("📌 Forcing RandomForest & extracting feature importance")

# Load data
data = load_diabetes(as_frame=True).frame

# Setup
s = setup(data, target="target", session_id=42, normalize=True, silent=True, html=False)

# Train + tune RF explicitly
model = create_model("rf")
tuned = tune_model(model)

# ✅ Feature importance handling (tree or linear fallback)
try:
    importances = tuned.feature_importances_
except:
    try:
        importances = np.abs(tuned.coef_)
    except:
        importances = None

# ✅ Get feature names from PyCaret API
try:
    cols = get_features()
except:
    cols = data.drop(columns=["target"]).columns

# Plot only if importances exist
if importances is not None:
    plt.figure(figsize=(6,4))
    plt.barh(cols, importances)
    plt.title("Feature Importance (RandomForest)")
    plt.tight_layout()
    plt.savefig("feature_importance_custom.png")
    plt.close()
    print("✅ Saved: feature_importance_custom.png")
else:
    print("⚠️ Model does not support feature importance")

EOF

```
