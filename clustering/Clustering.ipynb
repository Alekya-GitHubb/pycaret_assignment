```python
%%bash
micromamba run -n pycaret38 python - << 'EOF'
from pycaret.clustering import *
from sklearn.datasets import load_iris

df = load_iris(as_frame=True).frame
setup(data=df, normalize=True, session_id=42, silent=True, html=False)

model = create_model('kmeans')
plot_model(model, plot='cluster', save=True)
plot_model(model, plot='elbow', save=True)
plot_model(model, plot='silhouette', save=True)

print("✅ Plots saved: cluster.png, elbow.png, silhouette.png")

EOF
```

```python
%%bash
micromamba run -n pycaret38 python - << 'EOF'
from pycaret.clustering import *
from sklearn.datasets import load_iris

df = load_iris(as_frame=True).frame
setup(data=df, normalize=True, session_id=42, silent=True, html=False)
model = create_model('kmeans')

save_model(model, 'iris_kmeans_model')

print("✅ Model saved as iris_kmeans_model.pkl")

EOF
```

```python
%%bash
micromamba run -n pycaret38 python - << 'EOF'
from pycaret.clustering import *
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

print("📊 Generating cluster plot...")

# Load dataset
data = load_iris(as_frame=True).frame

# Setup clustering
s = setup(data, session_id=42, normalize=True, silent=True, html=False)

# Create KMeans model
model = create_model('kmeans')

# Plot clusters and save manually
plot_model(model, plot='cluster')
plt.savefig("cluster.png")
plt.close()

print("✅ cluster.png saved successfully!")
EOF
```
