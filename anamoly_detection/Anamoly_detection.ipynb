```python
%%bash
wget -q https://micro.mamba.pm/api/micromamba/linux-64/latest -O micromamba.tar.gz
tar -xf micromamba.tar.gz
```

```python
%%bash
./bin/micromamba create -y -n pycaret38 python=3.8
./bin/micromamba run -n pycaret38 pip install pycaret==2.3.5
echo "✅ PyCaret installed"
```

```python
%%bash
./bin/micromamba run -n pycaret38 python - << 'EOF'
from pycaret.anomaly import *
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

print("✅ Environment Loaded")

# Load small dataset (fast)
data = load_iris(as_frame=True).frame

# Setup
setup(data, session_id=42, silent=True, html=False)

# Train anomaly model (Isolation Forest)
model = create_model('iforest')

# Assign anomaly labels
preds = assign_model(model)

# ✅ Use correct column names (PyCaret 2.3.5 uses capital A)
preds[['Anomaly','Anomaly_Score']].to_csv("anomaly_predictions.csv", index=False)

# Save model
save_model(model, "isolation_forest_model")

# Plot anomaly scores
plt.figure(figsize=(6,4))
plt.scatter(range(len(preds)), preds['Anomaly_Score'], alpha=0.6)
plt.title("Anomaly Scores")
plt.xlabel("Row Index")
plt.ylabel("Score")
plt.savefig("anomaly_score_plot.png")
plt.close()
```
